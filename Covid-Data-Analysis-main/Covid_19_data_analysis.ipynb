```python
data.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace=True,axis=1) # here i am removing unecessary columns such as sno, Time, ConfirmedIndianNational,ConfirmedForeignNational.
```

```python
data
```

```python
# active cases number.
data['Active_cases']= data['Confirmed']-data['Deaths']-data['Cured']
```

```python
data['Active_cases']
```

```python
data.tail()
```

```python
statewise= pd.pivot_table(data,values=['Cured','Deaths','Confirmed'], index="State/UnionTerritory", aggfunc = max) # used to pivot the table and see no of cured,death,confirmed cases in all state/union territory.
```

```python
statewise
```

```python
# find the statewise recovery rate.
statewise['Recovery_rate']= (statewise['Cured']*100)/statewise['Confirmed']
```

```python
statewise # now we see recovery rate column is added to statewise table.
```

```python
#  now we will find statewise mortality rate and add it to the statewise table.
statewise['Mortality_rate']=(statewise['Deaths']*100)/statewise['Confirmed']
```

```python
statewise
```

```python
# now i will plot my pivot table with proper visuals.
statewise.style.background_gradient(cmap ="YlGnBu")
```

```python
# top 10 Active cases states.
top_10_active_case_states = data.groupby(by='State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
ax = sns.barplot(data=top_10_active_case_states.iloc[:10],y='Active_cases',x='State/UnionTerritory',linewidth=2,edgecolor='black')
plt.xlabel('State')
plt.ylabel('Active cases')
plt.title('Top 10 states with most active cases.')
plt.show()
```

```python
# top 10 states with max death cases.
top_10_death_case_states = data.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
ax = sns.barplot(data=top_10_death_case_states.iloc[:10],y='Deaths',x='State/UnionTerritory',linewidth=2,edgecolor='red')
plt.xlabel('State')
plt.ylabel('Death cases')
plt.title('Top 10 States with most death cases.')
plt.show()
```

```python
# Growth Trend of top 5 most affected state.
fig=plt.figure(figsize=(12,9))
ax=sns.lineplot(data=data[data['State/UnionTerritory'].isin(['Maharashtra','Tamil Nadu','Delhi','Telengana','Uttar Pradesh'])],x='Date',y='Active_cases',hue='State/UnionTerritory')
plt.title("Most affected Indian States",size=16)
```

```python
# remove row where state is india
vaccine_1=vaccine[vaccine['State']!="India"]
vaccine_1
```

```python
# top 5 most vaccinated states.
most_vaccination_states = vaccine_1.groupby(by='State').max()[['Total Individuals Vaccinated']].sort_values(by=['Total Individuals Vaccinated'],ascending=False).reset_index()
top_5_most_vaccinated=most_vaccination_states.iloc[:5]
top_5_most_vaccinated=top_5_most_vaccinated.set_index('State')
```
